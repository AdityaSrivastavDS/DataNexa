# ○Business Intelligence

•Business Intelligence refers to a method of extracting insights from raw data 

# ○Features of Business Intelligence

# 1. Quick Navigation

•After the cleaning of data it become very easy to navigate the required data from the data set

# 2. Instant Answers

•Refers to answers of questions for what we were searching for

•In simple terms it describes:

                              ♦what work for us
                              
                              ♦what not work for us
                              
                              ♦how did we go ahead and do stuffs

# 3. Decision Making

•After Instant answers now we have answers of all of our questions which helps us to understand the business better and then we can make better business decisions

# ○Phases of Business Intelligence 

# 1. IT Dependent

•Refers to data that is related to technical field 

•IT industry process these data and send it to Analyst

# 2. Analyst

•Analyst analyse the IT data and extract insights from it using Business Intelligence and different tools in which Data Visualization plays a crucial role

•Later analyst transfer those insights to end user

# 3. End User

•The insights then represented by end user to stake holders so that they can get a better understanding of business

# ○Data Visualization

•Data Visualization is intersection of design,information science and communication

•It refers to visualization or presentation of a set of information in more efficient way

# 1. Design

•It is first step in data visualization in which we form a design in which format we have to visualize our data and this done by deciding that which way is more suitable to present current information

# 2. Information Science 

•Information science is an academic field which is primarily concerned with analysis, collection, classification, manipulation, storage, retrieval, movement, dissemination, and protection of information.

# 3. Communication

•Communication plays a crucial role in data visualization as the report that has been prepared from raw data using BI,it needs to be explained in a efficient manner to stakeholders for better business understanding

# ○Benefits of Data Visualization

•Information in visualized format is better to understand in comparision to that of textual format

•Data Visualization explains the whole big amount of data in a very little which is more readible and understandable

•Imposes more impact on stakeholders

•Visualized data is more easy to explain 

◘Example :-

          For example give a book to a person for reading which consists of 3000 pages,that person will take more than a week to complete that book but still will not be capable of memorizing all things that he had been readen but when we will provide a movie format of that story book it will take him only 3-4 hours to finish it and after that also he will be able to recall all things which he has seen upto long time.So,this is how data visualization has a upper hand as it allows us to have better understanding of things .                  

# ○Power BI

•Stands for Business Intelligence

•It is a tool that is used for analysing data and extracting insights from it 

# ○Components of Power BI

# 1. Power Query

•This function of Power BI is used for cleaning of data that includes:

   -removing null values
   
   -deleting empty records etc.

# 2. Power Pivot

•This function of Power BI is used to connect or build relation between data from various data sources

# 3. Power View

•This function of Power BI is used for visualizing data

•It offers 30+ different charts to visualize data

•It allows to import data from various data sources

# 4. Power BI Service

•When we hav to share our prepared report all over the world then we use Power BI Service

# 5. Power Map

•This component of Power BI is used for visualization of geographic data.
•It includes visualization tools like heat map etc

♦Note :- Geographic data are types of data that are located at different geographical locations

# 6. Power BI Q n A

•This component of Power BI drill through data and get instant answers of questions for better understanding

•In simple terms it used to extract meta data from the data

# ○Power BI vs Power BI Service

•Power BI is a desktop application which is used for development while Power BI Service is a online service which used to publish prepared reports

# ○Power BI DAX

•DAX stands for Data Analysis Expression

•DAX is a collection of functions,operators and constant used in formulas,to return one or more values.

•It is a native formula or Query Language by Microsoft.

♦The DAX Expression contains:

•Measure is the name of new variable we wanted to create and it can have any name

•Function is the operation that we wanted to perform like COUNT,SUM etc.

•Then the name of the table on which we are currently working 

•And at last the column name on which we wanted to perform operation.

♦Context: 

        •It is the ability to create a function.
        
        •There are two type of constant in context:-
        
                                                    •Role Context
                                                    
                                                    •Filter Context


♦Note:- The DAX only and only works on relational data
